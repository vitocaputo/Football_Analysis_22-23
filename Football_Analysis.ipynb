```python
# Categorized the age column in 4 different groups
def age_group(a):
    if a <= 20:
        return 'under20'
    elif 20 < a <= 25:
        return '21-25'
    elif 25 < a <= 30:
        return '26-30'
    else:
        return 'over30'
```

```python
# Added the group column in the AP table
ap['Group_Age'] = ap['Age'].apply(age_group)
```

```python
# Added the group column in the AP table
fs['Group_Age'] = fs['Age'].apply(age_group)
```

```python
''' MINUTES PLAYED _ MATCHES PLAYED by AGE '''

age_min = fs.pivot_table(index="Group_Age", columns="League", values="Minutes_Played")
age_min=age_min.astype(int)
age_match = fs.pivot_table(index="Group_Age", columns="League", values="Matches_Played")
age_match=age_match.astype(int)

fig = px.imshow(age_min,
                labels=dict(x="Leagues", y="Age Groups", color="Minutes Played"),
                 y=['under20', '20-25', '26-30', 'over30'], aspect="auto", text_auto=True)
fig.update_layout(
    title={
        'text': "Minutes Played by Age Group",
        'y':0.95,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top'},
    paper_bgcolor="white")


fig1 = px.imshow(age_match,
                labels=dict(x="Leagues", y="Age Groups", color="Matches Played"),
                 y=['under20', '20-25', '26-30', 'over30'], aspect="auto", text_auto=True)
fig1.update_layout(
    title={
        'text': "Matches Played on average by Age Group",
        'y':0.95,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top'},
    paper_bgcolor="white")

fig.show()
fig1.show()
```

```python
fs["Color_Cards_Y"] = "yellow"
fs["Color_Cards_R"] = "red"
```

```python
''' CARDS BY LEAGUE violin plot'''

fig = go.Figure()

fig.add_trace(go.Violin(x=df_combined['N_Card'][ df_combined['Color_Card'] == 'yellow' ],
                        y=df_combined['League'][ df_combined['Color_Card'] == 'yellow' ],
                        legendgroup='Yellow', name='Yellow',
                        side='positive',
                        line_color='gold')
             )
fig.add_trace(go.Violin(x=df_combined['N_Card'][ df_combined['Color_Card'] == 'red' ],
                        y=df_combined['League'][ df_combined['Color_Card'] == 'red' ],
                        legendgroup='Red', name='Red',
                        side='positive',
                        line_color='tomato')
             )
fig.update_traces(meanline_visible=True)
fig.update_traces(orientation='h', side='positive', width=2, points=False)
fig.update_layout(template='plotly_white', title="Cards by League")
fig.show()
```
